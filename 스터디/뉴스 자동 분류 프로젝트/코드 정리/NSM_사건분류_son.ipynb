## Contents 단어추출

```python
from ckonlpy.tag import Twitter
twitter = Twitter()
```

```python
## 사전 계속 추가 feat.현호
add_dict = ['마우리시오', 
            '포체티노', 
            '아시안게임', 
            '그리고',]
twitter.add_dictionary(add_dict, 'Noun')
```

```python
with open('C:/Users/student/Desktop/new_sight/불용어사전.pickle', 'rb') as f:
    stop_words = pickle.load(f)
```

```python
#기본적인 전처리
df_news['content_text'] = df_news['content'].apply(lambda x: re.sub("[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]","",x))
df_news['content_text'] = df_news['content_text'].apply(lambda x: x.replace("\n",""))
df_news['content_text'] = df_news['content_text'].apply(lambda x: x.replace("손흥민",""))
df_news['content_text'] = df_news['content_text'].apply(lambda x: x.replace("토트넘",""))
df_news['content_text'] = df_news['content_text'].apply(lambda x: x[:-7]) #기자 이름 삭제
df_news['content_text'] = df_news['content_text'].apply(lambda x: twitter.nouns(x))
df_news['content_text'] = df_news['content_text'].apply(lambda x: [element for element in x if element not in stop_words ])
df_news['content_text'] = df_news['content_text'].apply(lambda x: [element for element in x if len(element) >1 ])
```

```python
noun_news = []
for n in range(len(df_news['content_text'])) :
    tmp_text = " ".join(df_news['content_text'][n])
    noun_news.append(tmp_text)
```

**TF-IDF 생성**<br>
*max_features* : 속성(단어)의 최대 개수. <br>
*min_df* : DF가 이 수치 미만이면 무시. 정수는 개수를 나타내고 실수는 문서의 비율을 의미함. <br> 
*max_df* : DF가 이 수치를 초과하면 무시. 정수는 개수를 나타내고 실수는 문서의 비율을 의미함. stopword를 걸러내려는 목적으로 설정함. <br> 

```python
vectorizer = TfidfVectorizer(min_df = 0.02, max_df = 0.7, stop_words = stop_words)
X = vectorizer.fit_transform(noun_news).toarray()
```

```python
vectorizer.get_feature_names()
```

```python
print(len(vectorizer.get_feature_names()))
```

```python
tfidf = 1 - np.round(pairwise_distances(X, metric="cosine"),3)
tfidf
```

```python
for row in range(tfidf.shape[0]) :
    for col in range(tfidf.shape[1]) :
        if row == col :
            continue
        elif row < col and tfidf[row,col] > 0.8 :
            print(f'{row}번 문서, {col}번 문서의 tf-idf문서 유사도 : {tfidf[row,col]}')
```
